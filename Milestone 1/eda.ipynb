```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cases_train = pd.read_csv('data/cases_2021_train.csv')
cases_test = pd.read_csv('data/cases_2021_test.csv')
location = pd.read_csv('data/location_2021.csv')
```

```python
def countNa(col):    
    count = col.isna().sum()
    print (col.name + ": "+ str(count))
```

```python
print("MISSING VALUES\nCASES_TRAIN_2021_________________________")
cases_train.apply(countNa)
print("\nCASES_TEST_2021_________________________")
cases_test.apply(countNa)
print("\nLOCATION_2021_________________________")
location.apply(countNa)
```

# CASES_TRAIN

```python
sns.set_style('darkgrid')
dims = (10, 15)
fig, ax = plt.subplots(figsize=dims)
plot = sns.histplot(y='country', data=cases_train, ax=ax);
plt.savefig("plots/task-1.3/train_country.png")
```

```python
new_df = cases_train[cases_train['country'] == "India"]
sns.set_style('darkgrid')
dims = (10, 15)
fig, ax = plt.subplots(figsize=dims)
plot = sns.histplot(y='province', data=new_df, ax=ax);
plt.savefig("plots/task-1.3/train_province.png")
```

```python
sns.set(rc = {'figure.figsize':(45,15)})
sns.scatterplot(data = cases_train, x='longitude', y='latitude', color='green');
plt.savefig("plots/task-1.3/train_lat_long.png", bbox_inches='tight')
```

```python
date = cases_train['date_confirmation'].str[3:10]
temp_data = cases_train.copy()
temp_data['date_confirmation'] = date
temp_data = temp_data.sort_values('date_confirmation')
```

```python
dims = (6, 6)
fig, ax = plt.subplots(figsize=dims)
sns.histplot(x = 'date_confirmation', data = temp_data, color='black', ax=ax);
plt.savefig("plots/task-1.3/train_date.png", bbox_inches='tight')
```

# CASES_TEST

```python
sns.set_style('darkgrid')
dims = (10, 15)
fig, ax = plt.subplots(figsize=dims)
plot = sns.histplot(y='country', data=cases_test, ax=ax);
plt.savefig("plots/task-1.3/test_country.png")
```

```python
new_df_test = cases_test[cases_test['country'] == "India"]
sns.set_style('darkgrid')
dims = (10, 15)
fig, ax = plt.subplots(figsize=dims)
plot = sns.histplot(y='province', data=new_df_test);
plt.savefig("plots/task-1.3/test_province.png")
```

```python
sns.set(rc = {'figure.figsize':(45,15)})
sns.scatterplot(data = cases_test, x='longitude', y='latitude', color='purple');
plt.savefig("plots/task-1.3/test_lat_long.png", bbox_inches='tight')
```

```python
date_test = cases_test['date_confirmation'].str[3:10]
temp_data_test = cases_test.copy()
temp_data_test['date_confirmation'] = date_test
temp_data_test = temp_data_test.sort_values('date_confirmation')
dims = (6, 6)
fig, ax = plt.subplots(figsize=dims)
sns.histplot(x = 'date_confirmation', data = temp_data_test, color='black', ax=ax);
plt.savefig("plots/task-1.3/test_date.png", bbox_inches='tight')
```
